```python
from random import choice
from time import sleep

# You have X guesses remaining
# You previously guessed: []
# BL_ANK word with guessed letters filled into blanks
# HANGMANPICS[x]

words = ('ant baboon badger bat bear beaver camel cat clam cobra cougar '
         'coyote crow deer dog donkey duck eagle ferret fox frog goat '
         'goose hawk lion lizard llama mole monkey moose mouse mule newt '
         'otter owl panda parrot pigeon python rabbit ram rat raven '
         'rhino salmon seal shark sheep skunk sloth snake spider '
         'stork swan tiger toad trout turkey turtle weasel whale wolf '
         'wombat zebra ').split()

# Hangman asci art
HANGMANPICS = ['''
  +---+
  |   |
      |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''', '''
  +---+
  |   |
  X   |
 /|\  |
 / \  |
      |
=========''']
victory = '''
       _      _                   
      (_)    | |                  
__   ___  ___| |_ ___  _ __ _   _ 
\ \ / / |/ __| __/ _ \| '__| | | |
 \ V /| | (__| || (_) | |  | |_| |
  \_/ |_|\___|\__\___/|_|   \__, |
                             __/ |
                            |___/ '''


def playGame():
    answer = choice(words)
    answerList = list(answer)
    guesses = 0
    attemp = ""
    used = []

    for letter in answer:
        attemp += "-"

    print(answer)

    while guesses <= 7:
        if attemp == answer:
            print("you win!")
            print(f"The word: {attemp}")
            print(victory)
            break
        elif guesses > 6:
            print("\n\nYou lost!")
            print(HANGMANPICS[guesses])
            print(f"The answer was: {answer}")
            break

        print("\n--Terminal Hangman--")
        print(f"you have used {guesses}/6 guesses")
        print(HANGMANPICS[guesses])
        print(f"Guess the word: {attemp}")
        print(f"You have guessed: {used}")
        guess = input("Guess a letter: ")
        sleep(.5)

        if guess in used:
            print(f"!!!!!\nYou havelready tried {guess}\n")
            continue

        if guess in answer:
            test = answerList.index(guess)

            indices = []
            for idx, value in enumerate(answer):
                if value == guess:
                    indices.append(idx)
            # print(indices)

            # print('test index', test)
            attempList = list(attemp)
            # print("attemptList: ", attempList)

            for index in indices:
                attempList[index] = guess

            attemp = "".join(attempList)

            print(f"\n{guess} is in the answer")
            used.append(guess)
            continue
        else:
            guesses += 1
            print(f"\n{guess} is not in the answer")

            used.append(guess)
            continue


game = input("Would you like to play hangman? y/n")
if game == "y":
    playGame()
else:
    print("See ya later!")


# Come back and make play again function
```
